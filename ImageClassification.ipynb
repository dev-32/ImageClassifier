```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# **Model 1(REsNet50)**

```python
IMG = '/content/sample_data/jon-koop-khYVyHiNZo0-unsplash.jpg'
```

```python
image = plt.imread(IMG)
plt.imshow(image)
```

# **Creating a Model**

```python
model_1 = tf.keras.applications.ResNet50()
```

# **Pre-processing Image**

```python

from tensorflow.keras.preprocessing import image
img1 = image.load_img(IMG, target_size=(224,224),color_mode='rgb',keep_aspect_ratio=True)
```

```python
resized_img1 = image.img_to_array(img1)
final_image1 = np.expand_dims(resized_img1,axis=0) ## Need four dimension
final_image1 = tf.keras.applications.resnet50.preprocess_input(final_image1)
```

# **Predicting from Model**

```python
prediction = model_1.predict(final_image1)
```

```python
final_image1.shape
```

**Decoding the predictions**

```python
from tensorflow.keras.applications import imagenet_utils
results_1 = imagenet_utils.decode_predictions(prediction)
```

# **Result**

```python
print(results_1)
```

```python
print(results_1[0][0][1])
```
